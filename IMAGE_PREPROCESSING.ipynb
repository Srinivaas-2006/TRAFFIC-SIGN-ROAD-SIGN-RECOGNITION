```python
import torch
import torchvision.transforms as transforms
from torchvision.datasets import GTSRB
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import random

# Image preprocessing steps
transform = transforms.Compose([
    transforms.Resize((64, 64)),          # Resize image
    transforms.ToTensor(),                # Convert image to tensor
    transforms.Normalize(                 # Normalize image
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Download dataset directly
train_dataset = GTSRB(
    root="/content/gtsrb_data",
    split="train",
    download=True,
    transform=transform
)

test_dataset = GTSRB(
    root="/content/gtsrb_data",
    split="test",
    download=True,
    transform=transform
)

# Create DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False
)

print("Image Preprocessing Completed Successfully")
print("Training images:", len(train_dataset))
print("Testing images:", len(test_dataset))
print("Image size: 64 x 64")
print("Batch size: 32")
print("Normalization: Applied")
print("Number of classes: 43")

# Function to display normalized images clearly
def denormalize(img):
    mean = torch.tensor([0.485, 0.456, 0.406]).view(3, 1, 1)
    std = torch.tensor([0.229, 0.224, 0.225]).view(3, 1, 1)
    img = img * std + mean
    return torch.clamp(img, 0, 1)

# Display preprocessed output images
plt.figure(figsize=(16, 9), dpi=240)

for i in range(12):
    index = random.randint(0, len(train_dataset) - 1)
    image, label = train_dataset[index]

    image = denormalize(image)
    image = image.permute(1, 2, 0).numpy()

    plt.subplot(3, 4, i + 1)
    plt.imshow(image)
    plt.title("Class: " + str(label), fontsize=12)
    plt.axis("off")

plt.suptitle("Preprocessed Traffic Sign Images", fontsize=20)
plt.tight_layout()
plt.savefig("preprocessed_traffic_sign_images_4k.png", dpi=240)
plt.show()

print("Preprocessed image output saved as preprocessed_traffic_sign_images_4k.png")
```
